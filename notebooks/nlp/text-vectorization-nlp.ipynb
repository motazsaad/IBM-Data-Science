Converting Score values into Posituve or Negative.

```python
pos_neg_data = data[data['Score']!=3]
def get_label (review_score):
    if review_score < 3:
        return 'negative'
    return 'positive'

pos_neg_data['label'] = data['Score'].apply(get_label)
```

Removing stopwords

# Stemming

```python
import re
snow = nltk.stem.SnowballStemmer('english')
def stemm_text(text):

    sent = text.lower()
    sent = re.sub('<.*?>','',sent)  # Remove html tag
    sent = re.sub(r'[?|!|\'|"|#]',r'',sent)
    sent = re.sub(r'[.|,|)|(|\|/]',r' ',sent)  # Remove punctuations

    words = [snow.stem(word) for word in sent.split() if word not in stopwords.words('english')]
    return ' '.join(words)
```

```python
final_data['stem'] = final_data['Text'].apply(stemm_text)
final_data
```

```python
def get_word_count(text):
    return len(text.split())
final_data['text WC'] = final_data['Text'].apply(get_word_count)
final_data['stem WC'] = final_data['stem'].apply(get_word_count)
final_data
```
